```python
import pandas as pd
```

```python
df = pd.read_csv("Countries-Europe.csv")
df = pd.DataFrame(df)

```

```python
for row in df.itertuples():
    if row.population >= 50000000:
        print(f"{row.Index} Ogromny kraj: {row.name}")
        print("Populacja: ", df.loc[row.Index, "population"])
        print("--------------------------------------------------")
    elif row.population >= 10000000 and row.population < 50000000:
        print(f"{row.Index} Duży kraj: {row.name}")
        print("Populacja: ", df.loc[row.Index, "population"])
        print("--------------------------------------------------")
    elif row.population >= 1000000 and row.population < 10000000:
        print(f"{row.Index} Średni kraj: {row.name}")
        print("Populacja: ", df.loc[row.Index, "population"])
        print("--------------------------------------------------")
    elif row.population < 1000000:
        print(f"{row.Index} Mały kraj: {row.name}")
        print("Populacja: ", df.loc[row.Index, "population"])
        print("--------------------------------------------------")


```
