```python
import pandas as pd
import matplotlib.pyplot as plt
import random

colors = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'pink', 'brown', 'gray', 'cyan']
x = random.choice(colors)


a = pd.read_csv('car_price_dataset.csv')

df = pd.DataFrame(a)

print(df.groupby('Year')['Engine_Size'].mean().plot(kind='line', figsize=(12,5), color=x))

```
